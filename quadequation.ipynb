```python
import math
import cmath
```

```python
a = float(input("a = "))
b = float(input("b = "))
c = float(input("c = "))
```

```python
d = b ** 2 - 4 * a * c
if d > 0:
    print((-b - math.sqrt(d)) / (2 * a), (-b + math.sqrt(d)) / (2 * a) )
elif d == 0:
    print(-b / (2 * a))
else:
    print((-b + cmath.sqrt(d)) / (2 * a), (-b - cmath.sqrt(d)) / (2 * a))
```
